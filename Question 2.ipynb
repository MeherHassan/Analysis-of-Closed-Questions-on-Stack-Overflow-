```python
import pandas as pd
df1 = pd.read_csv('posts.csv') 

df2 = pd.read_csv('PostHistory.csv')
df1.head(2)
```

```python
df2.head(2)
```

```python
out = (df2.merge(df1, left_on='Id', right_on='Id')
          .reindex(columns=['Id', 'PostHistoryTypeId', 'ClosedDate','Comment']))
```

```python
out = out[out['PostHistoryTypeId'] == 10]
out.head(50)
```

```python
import datetime as dt 
out['ClosedDate'] = pd.to_datetime(out['ClosedDate'])

out['MonthYear_value'] = out['ClosedDate'].dt.to_period('M')
```

```python
out.head(10)
```

```python
out = out[(out['MonthYear_value']>= "2008-08") & (out['MonthYear_value']<= "2018-08")]
out = out.sort_values(by='MonthYear_value')
out.head(50)
```

```python
x=out.groupby(['MonthYear_value','Comment'])['Comment'].count()

y=out.groupby(['MonthYear_value'])['Comment'].count()
r=((x/y)*100).round(2)
#print(r)
```

```python
df = pd.DataFrame(r)
#df.head(120)
pd.set_option('display.max_rows', df.shape[0]+1)
df
```
