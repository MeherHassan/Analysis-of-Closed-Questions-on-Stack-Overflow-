```python
import datetime as dt 

df2['CreationDate'] = pd.to_datetime(df2['CreationDate'])
df2['MonthYear'] = df2['CreationDate'].dt.to_period('M')
df2.head(2)
```

```python
df2 ['count']=df2.Text.str.count('Id')
df2.head(10)
```

```python
df = df2[['MonthYear','count']]
df.head(5)
```
