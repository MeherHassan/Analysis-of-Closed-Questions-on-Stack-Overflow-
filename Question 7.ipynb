```python
import pandas as pd
df1 = pd.read_csv('posts.csv')
df1.head(2)

```

```python
df2 = pd.read_csv('PostHistory.csv')
df2.head(2)
```

```python
dfinal = df1.merge(df2, on="Id")
dfinal.head(5)
```

```python
rslt_df = dfinal[dfinal['PostHistoryTypeId'] == 10]
rslt_df.head(5)
```

```python
df2 = rslt_df[['CreationDate_y', 'ClosedDate','Comment']]
df2.head(5)
```

```python
import datetime as dt 

df2['CreationDate_y'] = pd.to_datetime(dfinal['CreationDate_y'])

df2['ClosedDate'] = pd.to_datetime(dfinal['ClosedDate'])
```

```python
df2['time_delta'] = (df2.ClosedDate - df2.CreationDate_y)

#df2['tot_hour_diff'] = (df2.ClosedDate - df2.CreationDate_y) / pd.Timedelta(hours=1)
df2['tot_mins_diff'] = (df2.ClosedDate - df2.CreationDate_y) / pd.Timedelta(minutes=1)
```

```python
df2
```

```python
df2 = df2[['Comment', 'tot_mins_diff']]
df2.head(5)
```

```python
df2.boxplot(column=['tot_mins_diff' ], by = ['Comment'])
```

```python
x=df2.groupby(['Comment','tot_mins_diff'])['tot_mins_diff'].count()

y=df2.groupby(['Comment'])['tot_mins_diff'].count()
r=((x/y)*100).round(2)

dff = pd.DataFrame(r)
#df.head(120)
pd.set_option('display.max_rows', dff.shape[0]+1)
dff
```
